son los 10 paises con mayor valor per capita de consumo de energia electrica por persona (kWH/person) en el año 2017

```python
top=df[(df["Year"]==2022) & (~df["Code"].isna())].sort_values("Primary energy consumption per capita (kWh/person)",ascending=False).head(10)
top
```

En nuestra presentacion, se agregara a pero para poder tener una comparativa contra los 10 primeros paises

```python
top_values=top['Entity'].values
```

```python
top_values
```

```python
top_values_PER  = (list(top_values))
```

```python
top_values_PER.append("Peru")
```

```python
top_values_PER
```

```python
df_top_PER= df[df['Entity'].isin(top_values_PER)]
df_top_PER
```

```python
df_reg=df[(df["Code"].isna()) | (df['Entity']=="Peru")]
```

```python
import plotly.express as px
df = px.data.gapminder()
fig = px.line(df_top_PER, x="Year", y="Primary energy consumption per capita (kWh/person)", color="Entity", hover_name="Entity",
        line_shape="spline", render_mode="svg", log_y=True)
fig.update_layout(
    title="Top 10 of countries and Perú with the highest consumption per capita ")
fig.show()
```

Podemos observar que los pasises con mayor consumi per capita de energia anual superan hasta más de 10 veces el consumo energetico de los peruano, generalmente se puede asociar el consumo energetico por habitante con el nivel de de insdustrializacion del paise, sin embargo no siempre es asi, podemos ver que paises como canada e islandia que estan cerca a los hemisferios, requerien una mayor cantidad de energia debido a su consumo de energia para aclimatar sus ambientes, ahora analizamos el crecimiento exponencial que tuvo Oman en el ciclo pasado, que es un pais actualmente petrolero pero que el ciclo pasado su mayor fuente de ingreso era la agricultura, lo mismo sudece con otros paises del medio oriente como lo son los paises como Emiratos Arabes Unidos y Qatar.
